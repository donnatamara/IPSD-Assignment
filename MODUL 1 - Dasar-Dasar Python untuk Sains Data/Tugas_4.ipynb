### Soal 4: Studi Kasus Pengelolaan Data

**Soal**: Anda diberikan file CSV berisi data nilai ujian mahasiswa. Tugas Anda adalah menulis sebuah program yang:
1. Membaca file CSV dan menyimpan datanya ke dalam dictionary.
2. Menghitung rata-rata nilai tiap mahasiswa.
3. Menampilkan mahasiswa dengan nilai tertinggi dan terendah.

```python
import pandas as pd
from google.colab import files
import io

# Upload dan membaca file CSV
up = files.upload()

# Mendapatkan nama file yang diunggah
nama_file = list(up.keys())[0]

# Membaca file CSV menjadi DataFrame
df = pd.read_csv(io.StringIO(up[nama_file].decode('utf-8')))
print(df)

# Mengubah DataFrame menjadi dictionary dengan Nama Siswa sebagai kunci
dictionary = df.set_index('Nama Siswa').T.to_dict(orient='list')
print(dictionary)

# Fungsi untuk menghitung rata-rata nilai tiap siswa
def hitung_rata_rata(data_mahasiswa):
    rata_rata_nilai = {}
    for nama, nilai in data_mahasiswa.items():
        rata_rata = sum(nilai) / len(nilai)
        rata_rata_nilai[nama] = rata_rata
    return rata_rata_nilai

# Menghitung rata-rata keseluruhan dari semua nilai
def hitung_rata_rata_keseluruhan(data_mahasiswa):
    total_nilai = 0
    total_data = 0
    for nilai in data_mahasiswa.values():
        total_nilai += sum(nilai)
        total_data += len(nilai)
    rata_rata_keseluruhan = total_nilai / total_data
    return rata_rata_keseluruhan

rata_rata_keseluruhan = hitung_rata_rata_keseluruhan(dictionary)

# Menghitung rata-rata
rata_rata_nilai = hitung_rata_rata(dictionary)

# Menentukan siswa dengan nilai tertinggi dan terendah
mahasiswa_nilai_tertinggi = max(rata_rata_nilai, key=rata_rata_nilai.get)
mahasiswa_nilai_terendah = min(rata_rata_nilai, key=rata_rata_nilai.get)

# Menampilkan hasil akhir
print("Rata-rata nilai setiap siswa:")
for nama, rata_rata in rata_rata_nilai.items():
    print(f"{nama}: {rata_rata:.2f}")

print(f"\nSiswa dengan nilai tertinggi: {mahasiswa_nilai_tertinggi} ({rata_rata_nilai[mahasiswa_nilai_tertinggi]:.2f})")
print(f"Siswa dengan nilai terendah: {mahasiswa_nilai_terendah} ({rata_rata_nilai[mahasiswa_nilai_terendah]:.2f})")
print(f"Rata-rata keseluruhan dari semua nilai siswa: {rata_rata_keseluruhan:.2f}")
```
